```python
import numpy as np
import pandas as pd

# 1. Lists
list_data = [1, 2, 3, 4, 5]
print("List data:", list_data)
print("Type of list data:", type(list_data))

# 2. Arrays
array1 = np.array([1, 2, 3, 4, 5])
array2 = np.array([5, 4, 3, 2, 1])
print("Array data 1:", array1)
print("Array data 2:", array2)
print("Type of array data:", type(array1))

# 3. Mathematical Operations
print("Multiplication:", array1 * array2)
print("Division:", array1 / array2)
print("Power (array1^2):", np.power(array1, 2))

# 4. Combine text with Numpy operation
print(f"Addition of Two: array1 + array2 = {array1 + array2}")

# 5. Implement np.sin(), np.log(), np.log2(), np.exp()
print("Sine of array1:", np.sin(array1))
print("Log of array1:", np.log(array1))  # Note: log(0) is undefined
print("Log2 of array1:", np.log2(array1))
print("Exponential of array1:", np.exp(array1))
```

```python
import matplotlib.pyplot as plt

# Dataset for Health Care
people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

# 1. Scatter Plot
plt.figure(figsize=(8, 5))
plt.scatter(age, height, color='blue')
plt.title("Age vs Height")
plt.xlabel("Age (years)")
plt.ylabel("Height (cm)")
plt.show()

# 2. Bar Chart
plt.figure(figsize=(8, 5))
plt.bar(people, weight, color='green')
plt.title("Weight of Individuals")
plt.xlabel("Person")
plt.ylabel("Weight (kg)")
plt.show()

# 3. Histogram
plt.figure(figsize=(8, 5))
plt.hist(age, bins=5, color='purple')
plt.title("Age Distribution")
plt.xlabel("Age (years)")
plt.ylabel("Frequency")
plt.show()

# Explanation:
# - Scatter Plot: Shows the relationship between age and height. As age increases, height tends to increase.
# - Bar Chart: Displays the weight for each person, useful for visualizing individual data points.
# - Histogram: Shows the distribution of ages, helping to see how ages are spread across the sample.
```
